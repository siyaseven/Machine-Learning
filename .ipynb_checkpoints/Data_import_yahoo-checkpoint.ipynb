```python
months31 = [1,3,5,7,8,10,12]
months30 = [4,6,9,11]
Date = []
year = int(input("Enter the Year:"))
for j in range(1,13):
    if j in months31:
        for i in range(1,32):
            date_obj = datetime.datetime(year, j, i, 18, 36, 28, 151012)
            Date.append(date_obj)
    elif j in months30:
        for i in range(1,31):
            date_obj = datetime.datetime(year, j, i, 18, 36, 28, 151012)
            Date.append(date_obj)
    else:
        if year % 4 == 0:
            for i in range(1,30):
                date_obj = datetime.datetime(year, j, i, 18, 36, 28, 151012)
                Date.append(date_obj)
        else:
            for i in range(1,29):
                date_obj = datetime.datetime(year, j, i, 18, 36, 28, 151012)
                Date.append(date_obj)
```

```python
fx_rates = []
for i in Date:
    fx_rates.append(c.get_rate("USD","ZAR", i))
```

```python

fig, ax = plt.subplots(figsize=(16,9))

ax.plot(Date, fx_rates, label='USD.ZAR')
#ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling')
#ax.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
```

```python
def get_fx_rate(BaseCCY, SecondaryCCY, StartDate,EndDate):
    date = pd.date_range(start=StartDate, end=EndDate)
    Date = []
    fx_rates = []
    for i in date:
        Date.append(datetime.fromisoformat(str(i)))
    for i in Date:
        fx_rates.append(c.get_rate(BaseCCY,SecondaryCCY, i))
    return  fx_rates
    
```

```python
class Get_FX_Rates():
    def __init__(self,baseCCY, secondaryCCY, startDate, endDate):
        self.baseCCY = baseCCY
        self.secondaryCCY = secondaryCCY
        self.startDate = startDate
        self.endDate = endDate
        
    def get_date(startDate,endDate):
        date = pd.date_range(start=startDate, end=endDate)
        return  [datetime.fromisoformat(str(i)) for i in date]
    
    @classmethod
    def get_fx_quotes(cls,baseCCY,secondaryCCY,startDate,endDate):
        c = CurrencyRates()
        return [c.get_rate(baseCCY,secondaryCCY, i) for i in cls.get_date(startDate,endDate)]
    
    @classmethod
    def fx_rates_against_date(cls,baseCCY,secondaryCCY,startDate,endDate):
       
        fig, ax = plt.subplots(figsize=(16,9))
                              
        ax.plot(cls.get_date(startDate,endDate), cls.get_fx_quotes(baseCCY,secondaryCCY,startDate,endDate), label= baseCCY + secondaryCCY)
        ax.set_xlabel('Date')
        ax.set_ylabel('Adjusted closing price')
        ax.legend()l;
        
```

# Now Downloading stock data using Yahoofinancials

```python
yahoo_financials = YahooFinancials('TSLA')

data = yahoo_financials.get_historical_price_data(start_date='2000-01-01', 
                                                  end_date='2019-12-31', 
                                                  time_interval='weekly')

tsla_df = pd.DataFrame(data['TSLA']['prices'])
tsla_df = tsla_df.drop('date', axis=1).set_index('formatted_date')
tsla_df.head()
```
